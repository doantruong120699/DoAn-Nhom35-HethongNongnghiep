```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('Danang11thang4-9thang6.csv')
dataset
```

```python
def change_time(dataset):
    time_data = dataset['timestamp']
    time = []
    for i in range (len(time_data)):
        t = time_data[i].split('T', 1)
        time.append((t[1]))
    return time

dataset['timestamp'] = change_time(dataset)
```

```python
dataset
```

```python
data = pd.get_dummies(dataset[['timestamp', 'Temperature', 'Relative Humidity', 'Mean Sea Level Pressure', 'Wind Speed', 'Precipitation Total']])
data
```

```python
def getdata(data):
    data_output = data[['Precipitation Total']].values
    data_input = data[['Temperature', 'Relative Humidity', 'Mean Sea Level Pressure', 'Wind Speed', 'timestamp_0000', 'timestamp_0100', 'timestamp_0200', 'timestamp_0300', 'timestamp_0400', 'timestamp_0500', 'timestamp_0600', 'timestamp_0700', 'timestamp_0800', 'timestamp_0900', 'timestamp_1000', 'timestamp_1100', 'timestamp_1200', 'timestamp_1300', 'timestamp_1400', 'timestamp_1500', 'timestamp_1600', 'timestamp_1700', 'timestamp_1800', 'timestamp_1900', 'timestamp_2000', 'timestamp_2100', 'timestamp_2200', 'timestamp_2300']].values
    return data_output, data_input

data_output, data_input = getdata(data)
data_output.shape, data_input.shape
```

```python
y =data_output[data_output>0]
print("Số giờ mưa: ",y.shape[0] ,"/", data_output.shape[0])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(data_input, data_output, test_size=0.2, random_state=0)
```

```python
yy = y_train[y_train>0]
print("Số giờ mưa tập train: ",yy.shape[0] ,"/", y_train.shape[0])
```

```python
yyy = y_val[y_val>0]
print("Số giờ mưa tập test: ",yyy.shape[0] ,"/", y_val.shape[0])
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_val = sc.transform(X_val)
```

```python
import keras 
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.optimizers import SGD
```

```python
unit_input = 28
unit_hidden = 20
unit_output = 1
```

```python
def build_model(unit_input, unit_hidden, unit_output):
    model = Sequential()
    # Add input layer
    model.add(keras.layers.Dense(unit_hidden, activation = 'relu', input_shape=(unit_input,)))
    # Add hidden layers
    model.add(keras.layers.Dense(unit_hidden, activation = 'relu'))
    # Add output layers
    model.add(keras.layers.Dense(unit_output, activation = 'linear'))
    return model
```

```python
model = build_model(unit_input, unit_hidden, unit_output)
# model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.compile(optimizer='adam', loss='mse')
```

```python
train_history = model.fit(X_train, y_train, batch_size=32, nb_epoch=1000)
```

```python
y_val_pred = model.predict(X_val)
mse_val = ((y_val - y_val_pred) ** 2).mean()
mse_val
```

```python
from keras.models import model_from_json
from keras.models import load_model
```

```python
# serialize model to JSON
#  the keras model which is trained is defined as 'model' in this example
model_json = model.to_json()
with open("model_rain.json", "w") as json_file:
    json_file.write(model_json)

# serialize weights to HDF5
model.save_weights("model_rain.h5")
```

```python
# load json and create model
json_file = open('model_rain.json', 'r')

loaded_model_json = json_file.read()
json_file.close()

loaded_model = model_from_json(loaded_model_json)

# load weights into new model
loaded_model.load_weights("model_rain.h5")
print("Loaded model from disk")

loaded_model.save('model_rain.hdf5')
model_new=load_model('model_rain.hdf5')
```
